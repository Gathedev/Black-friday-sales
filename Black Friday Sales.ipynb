# Data Cleaning:

```python
data.dtypes
```

```python
#filling the null in rating_count column by the mode value of the column as it's data type is object

data['rating_count'] = data['rating_count'].fillna(data['rating_count'].mode(0))
```

```python
data.isnull().sum()
```

```python
# checking for the outliers in the columns of interest in the dataset

data.columns
```

****Amongst all the features in the dataset there are certain features that will affect the customer rating of the product, we will consider only those columns and form a new dataframe named df with these features. We will check outliers and here after perform all the analysis on these features only.****

```python
df=data[['category', 'discounted_price','actual_price', 'discount_percentage', 'rating']]
```

```python
df.head()
```

***Now we are unable to check the outliers in the features of interest as there data type is object. So we will clean the data first and then check outliers.***

```python
def remove_symbol(var):
    s=var[1:]
    return s.replace(',','')
```

***Created a function named remove_symbol to remove the symbols from the features so that our features could be treated as integer or float and not as strings. This will make the data analysis easier.***

```python
df['discounted_price']=df['discounted_price'].apply(remove_symbol)
```

```python
df['actual_price']=df['actual_price'].apply(remove_symbol)
```

```python
df['discount_percentage']=df['discount_percentage'].apply(lambda x:x[:-1])
```

```python
df.head()
```

```python
df.dtypes
```

***Even though we have removed the symbols the features are still in object data type. So to convert its data type astyp function is used as follows***

```python
df['discounted_price']=df['discounted_price'].astype(float)
df['actual_price']=df['actual_price'].astype(float)
df['discount_percentage']=df['discount_percentage'].astype(float)
```

```python
df.dtypes
```

```python
#checking the outliers in the features

variables = ['discounted_price','actual_price', 'discount_percentage']

i=1
for var in variables:
    plt.figure(i)
    sns.boxplot(df[var])
    i += 1
    if(i==1):
        plt.title('Disconted price')
    if(i==2):
        plt.title('Actual price')
    if(i==3):
        plt.title('Discount percentage')
```

***From above box plots we can see that the actaul-price and discounted_price do have outliers in them***
